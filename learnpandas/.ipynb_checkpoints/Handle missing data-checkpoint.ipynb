```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(5, 3), index=['a', 'c', 'e', 'f', 'h'],
                 columns=['one', 'two', 'three'])
df['four']='bar'
df['five']= df['one'] > 0
df['date']= pd.Timestamp('2018-01-01')
df2 = df.reindex(['a','b', 'c','d', 'e', 'f','g', 'h'])

df2
```

```python
## NaN is the default marker of missing number, text or boolean, NaT is for missing datetime
## How to detect missing value,
## pandas provides two methods, isna() and notna()
df2.isna()
```

```python
df2.notna()
```

```python
## cleaning na data
#1. drop na value, dropna() has a parameter axis which indicates either colums(axis = 1)
# or rows(axis = 0) will be dropped, the default value of axis is 0
df2.dropna()
```

```python
# drop colums with missing data
df2.dropna(axis = 1)
```

```python
## fill na data
df2.fillna(0)
```

```python
## We can fill data with respect to their column property
df2.fillna(value = {'one': 0, 'four': 'missing', 'five': False, 'date': pd.Timestamp('2000-01-01')})
```

```python
## Remind that fillna() will not replace na in orignial dataframe, in other words, the df2 remains the same after all the above operation
df2
```

```python
# If you want it to be permanantely changed, add parameter inplace = True
df2.fillna(value = 0, inplace = True)
df2
```
