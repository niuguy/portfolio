```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(5,4), index = ['a', 'b', 'c', 'd', 'e'], columns= ['A', 'B', 'C', 'D'])

df
```

```python
# Using [] for columns selection
df[['A', 'B']]
```

```python
# Range slicing
df[1:3] # or df['b':'c']
```

```python
# .loc for rows and colums selection
df.loc['a':'c', ['A','B']]
```

```python
# select all columns
df.loc['a':'c', :]
```

```python
# select with boolean

df[df['A']>0.2]
```

```python
# same result select by callable(lambda)

df[lambda df: df['A']>0.2]
```
